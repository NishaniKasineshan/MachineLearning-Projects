satisfaction_level: Level of satisfaction {0–1}.

```python
#Reading data
df=pd.read_csv("HR_comma_sep.csv")
df.head(3)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle = df
dfle.Department = le.fit_transform(dfle.Department)
dfle
```

```python
dfle = df
dfle.salary = le.fit_transform(dfle.salary)
dfle
```

```python
dfle.drop('Department',axis="columns")
```

```python
X = dfle[['satisfaction_level',	'last_evaluation',	'number_project',	'average_montly_hours',	'time_spend_company',	'Work_accident',	'promotion_last_5years','salary']]
```

```python
Y=dfle.left
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler #dividing all values by standard deviation
from sklearn.pipeline import make_pipeline

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
pipe = make_pipeline(StandardScaler(), LogisticRegression())
pipe.fit(X_train, Y_train)
```

```python
X_test
```

```python
pipe.predict(X_test)
```

```python
pipe.score(X_test, Y_test) 
```

```python
subdf = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
subdf.head()
```

Tackle salary dummy variable

Salary has all text data. It needs to be converted to numbers and we will use dummy variable for that. 

```python
salary_dummies = pd.get_dummies(subdf.salary, prefix="salary")
```

```python
df_with_dummies = pd.concat([subdf,salary_dummies],axis='columns')
```

```python
df_with_dummies.head()
```

Now we need to remove salary column which is text data. It is already replaced by dummy variables so we can safely remove it

```python
df_with_dummies.drop('salary',axis='columns',inplace=True)
df_with_dummies.head()
```

```python
X = df_with_dummies
X.head()
```

```python
y = df.left
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.3)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
#Accuracy of the model
model.score(X_test,y_test)
```
